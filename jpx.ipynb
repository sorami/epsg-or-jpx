# 東証 上場銘柄一覧

ソース: [その他統計資料 | 日本取引所グループ](https://www.jpx.co.jp/markets/statistics-equities/misc/01.html)


```python
import pandas as pd
```

## Excel データの読み込み


```python
df = pd.read_excel("./data/jpx/data_j.xls")
df
```

## データの確認


```python
# コードが重複していないことを確認
df[df.duplicated(subset="コード", keep=False)]
```

```python
# コードが4桁でない銘柄
df["code_len"] = df["コード"].apply(lambda x: len(str(x)))
df[df["code_len"] != 4]
```

### 数字以外のコード


```python
df_alpha = df[~df["コード"].apply(lambda x: isinstance(x, int))]
print(df_alpha.shape)
df_alpha
```

### 業種


```python
df["33業種区分"].value_counts()
```

```python
# 業種
df["17業種区分"].value_counts()
```

## データの整理


```python
# 「銘柄名」文字列の正規化
df["銘柄名"] = df["銘柄名"].str.normalize("NFKC")
df.head()
```

```python
jpx_codes = {}

for _, row in df.iterrows():
    d = row.to_dict()

    code = str(d.pop("コード"))
    name = d.pop("銘柄名")
    market = d.pop("市場・商品区分")
    industry = d.pop("33業種区分")

    jpx_codes[code] = {
        "name": name,
        "market": market,
        "industry": industry,
    }

jpx_codes
```

## 出力


```python
import json

with open("./data/jpx.json", "w") as f:
    json.dump(jpx_codes, f, ensure_ascii=False, indent=4)
```
